```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,50)
```

```python
f = lambda x:x**2
h = lambda x:x**3
```

$f(-x) =  f(x) : par$ \
$f(-x) = -f(x) : impar$ \
paridade  de funções: importante para machine learning

```python


plt.scatter(x,f(x))
plt.scatter(x,h(x))
plt.grid(True)
plt.show()
```

Função sigmoid \
$\large \frac{1}{1+e^{x}}$ 

```python
o = lambda x: ( 1 )/ ( 1+math.exp(-x) )
```

```python
def sigmoid(x):return ( 1 )/ ( 1+math.exp(-x) )
```

```python
fx = []
for i in x:
    fx.append(sigmoid(i))
```
